```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
embeddings = model.encode(sentences)

print(embeddings)
```

```python
df = pd.read_csv('sejong1000.csv')

df.head()
```

```python
df = df[~df['A'].isna()]

df.head()
```

```python
df.loc[0, 'Q']
```

```python
model.encode(df.loc[0, 'Q'])
```

```python
df['embedding'] = pd.Series([[]] * len(df)) # dummy

df['embedding'] = df['Q'].map(lambda x: list(model.encode(x)))

df.head()
```

```python
df.to_csv('sejong1000.csv', index=False)
```

```python
text = '몇번째 왕이세요?'

embedding = model.encode(text)
```

```python
df['distance'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())

df.head()
```

```python
answer = df.loc[df['distance'].idxmax()]

print('label', answer['label'])
print('유사한 질문', answer['Q'])
print('챗봇 답변', answer['A'])
print('유사도', answer['distance'])
```
